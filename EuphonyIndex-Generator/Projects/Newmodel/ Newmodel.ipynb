```r
library(xgboost)
library(dplyr)
library(data.table)
library(Matrix)
```

```r
odd.number <- 2*(1:113)
odd.data <- dat[odd.number,]
even.data <- dat[-odd.number,]
dat.data <- rbind(odd.data, even.data)
dat.data
```

```r
model.data <- dat.data[1:113,] #学習データ
pre.data <- dat.data[114:226,] #評価データ
#data.table型へ変換
model.data.dt <- data.table(model.data,keep.rownames=F)
pre.data.dt <- data.table(pre.data,keep.rownames=F)
#data.table型からxgb.dgCMatrixへ変換
model.data.mx <- sparse.model.matrix(natorcong~.,data=model.data.dt)
model.data.dm <- xgb.DMatrix(model.data.mx,label=model.data.dt[,natorcong])

pre.data.mx <- sparse.model.matrix(natorcong~.,pre.data.dt)
pre.data.dm <- xgb.DMatrix(pre.data.mx,label=pre.data.dt$natorcong)
```

```r
#XGBoostパラメータ(今回は全てデフォルト値を採用)
params <- list(
  eta=0.3,
  gamma=0,
  max_depth=6,
  min_child_weight=1,
  max_delta_step=0,
  subsample=1,
  colsample_bytree=1,
  colsample_bylevel=1,
  lambda=1,
  alpha=0
)

#XGBoost
xgb.result <- xgb.train(params=params, data=model.data.dm, label=model.data.dt$natorcong, objective="binary:logistic", eval_metric="auc", nfold=5, booster="gbtree", nrounds=1000, verbose=1)

#評価
pred <- predict(object=xgb.result,newdata=pre.data.dm)

for(i in 1:length(pred)){
  if(pred[i]==0) {pred[i]=0}
  else if (pred[i]==1) {pred[i]=1}
}
```

```r
result <- t(rbind(pre.data$natorcong, pred))
colnames(result) <- c("natorcong", "pred")
result <- transform(result, "残差"=(pre.data$natorcong-pred))
```

```r
dat3 <- dat2[-odd.number]
E <- t(rbind("natorcong"=result$natorcong, "pred"=dat3, "残差"=(result$natorcong-dat3)))
result <- rbind(transform(result, Species="Newmodel"), transform(E, Species="Euphony"))
result
```

```r
adaboosting <- function(object, data, iter){ #一列目に応答数
    true_data <- data[,1]
    D_1 <- c()
    for (i in 1:nrow(data)){
        D_1 <- append(D_1, 1/nrow(data))
    }
    D_m <- matrix( , , nrow(data))
    D_m <- rbind(D_m, D_1)
    h_m <- list()
    Alpha_m <- c()
    for (i in 1:iter){
        dat0 <- data[,-1]
        dat <- cbind(object=data[,1], dat0*D_m[i+1,])
        pred_dat <- cbind(object=data[,1], dat0)
        h_m <- append(h_m, list(glm(dat$object~., data=dat, family=gaussian())))
        pred <- predict(object=h_m[[i]], newdata=pred_dat)
        pred[pred>=0.5] <- 1
        pred[pred<0.5] <- 0
        E_m <- c()
        d_m <- c()
        for (j in 1:nrow(data)){
            if (pred[j] == true_data[j]){
                E_m <- append(E_m, 0)
                d_m <- append(d_m, D_m[i+1,j])
            } else {
                E_m <- append(E_m, D_m[i+1,j])
                d_m <- append(d_m, D_m[i+1,j])
            }
        }
        E_m <- sum(E_m)/sum(d_m)
        Alpha_m <- append(Alpha_m, log((1-E_m)/E_m))
        Alpha_m[is.infinite(Alpha_m)] <- 0
        d_m <- c()
        for (j in 1:nrow(data)){
            if (pred[j] == true_data[j]){
                d_m <- append(d_m, D_m[i+1,j])
            } else {
                d_m <- append(d_m, D_m[i+1,j]*exp(Alpha_m[i]))
            }
        }
        D_m <- rbind(D_m, d_m)
    }
    return (c(h_m, Alpha_m)) 
}
```

```r
odd.data$natorcong <- as.numeric(as.character(odd.data$natorcong))
even.data$natorcong <- as.numeric(as.character(even.data$natorcong))
even.data2 <- even.data[,-c(7,8)]
odd.data2 <- odd.data[,-c(7,8)]
o.ada <- adaboosting(natorcong, odd.data2, 8)
o.ada
```
